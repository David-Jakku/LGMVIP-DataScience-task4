```python
import cv2
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
img=cv2.imread('F:\DATA\jonsnow.jpg')
plt.imshow(img)
```

```python
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
plt.axis(False)
plt.show()
```

```python
grey_scale_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(grey_scale_image)
```

```python
inv_img=cv2.bitwise_not(grey_scale_image)
plt.imshow(inv_img)
```

```python
blur_img=cv2.GaussianBlur(inv_img,(111,111),0)
plt.imshow(blur_img)
```

```python
inv_blur_img=cv2.bitwise_not(blur_img)
plt.imshow(inv_blur_img)
```

```python
sketch_img=cv2.divide(grey_scale_image,inv_blur_img,scale=256.0)
plt.imshow(sketch_img)
```

```python
cv2.imwrite("sketch.png",sketch_img)
```

```python
plt.figure(figsize=(8,8))
img2=cv2.cvtColor(sketch_img,cv2.COLOR_RGB2BGR)
plt.imshow(img2)
plt.axis(False)
plt.show()
```
